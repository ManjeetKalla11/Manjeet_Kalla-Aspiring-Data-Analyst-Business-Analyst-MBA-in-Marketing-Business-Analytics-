```python
# Trend Analysis: Number of Incidents Per Year
trend_data = df.groupby('iyear')['eventid'].count()

plt.figure(figsize=(12, 6))
sns.lineplot(x=trend_data.index, y=trend_data.values, marker='o')
plt.title('Trend of Terrorist Incidents Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Incidents')
plt.grid()
plt.show()

#Save image in png format
plt.savefig('trend_of_incidents_over_years.png')
```

```python
# Group data by region and count the number of incidents
region_data = df['region_txt'].value_counts().head(10)

# Plot the top 10 regions affected by terrorism
plt.figure(figsize=(12, 6))
sns.barplot(x=region_data.values, y=region_data.index, palette='viridis')
plt.title('Top 10 Regions Affected by Terrorism')
plt.xlabel('Number of Incidents')
plt.ylabel('Region')
plt.grid(axis='x')
plt.show()

#Save image in png format
plt.savefig('top_10_regions_affected.png')
```

```python


# Group and count attack types
attack_types = df['attacktype1_txt'].value_counts()

# Combine smaller categories into 'Other'
threshold = 0.02 * attack_types.sum()  # Define 2% threshold
attack_types_combined = attack_types[attack_types >= threshold]
other_count = attack_types[attack_types < threshold].sum()
attack_types_combined['Other'] = other_count

# Create a pie chart
plt.figure(figsize=(10, 10))
explode = [0.1 if i == 0 else 0 for i in range(len(attack_types_combined))]  # Explode the largest slice

plt.pie(
    attack_types_combined.values,
    labels=attack_types_combined.index,
    autopct='%1.1f%%',
    startangle=140,
    explode=explode,
    colors=sns.color_palette('pastel')
)

# Add title and show the chart
plt.title('Distribution of Attack Types')
plt.show()


#Save image in png format
plt.savefig('distribution_of_attack_types.png')
```

```python
# Horizontal Bar Chart: Perpetrator Groups
perpetrators = df['gname'].value_counts().head(10)
perpetrators = perpetrators[perpetrators.index != 'Unknown']

plt.figure(figsize=(12, 6))
sns.barplot(x=perpetrators.values, y=perpetrators.index, palette='coolwarm', orient='h')
plt.title('Top 10 Perpetrator Groups')
plt.xlabel('Number of Incidents')
plt.ylabel('Group Name')
plt.grid(axis='x')
plt.show()

#Save image in png format
plt.savefig('top_10_perpetrator_groups.png')
```

```python
# Calculate total casualties (killed + wounded)
df['total_casualties'] = df['nkill'] + df['nwound']

# Identify the top 10 incidents with the highest casualties
highest_casualties = df.nlargest(10, 'total_casualties')[['eventid', 'iyear', 'country_txt', 'region_txt', 'attacktype1_txt', 'total_casualties']]

# Plot the top 10 incidents with the highest casualties
plt.figure(figsize=(12, 8))
sns.barplot(
    x=highest_casualties['total_casualties'],
    y=highest_casualties['eventid'].astype(str),  # Use event ID as y-axis labels
    palette='rocket'
)

# Add labels and title
plt.title('Top 10 Incidents with Highest Casualties')
plt.xlabel('Total Casualties (Killed + Wounded)')
plt.ylabel('Event ID')
plt.grid(axis='x')
plt.show()

# Display the detailed table for reference
print(highest_casualties)


#Save image in png format
plt.savefig('top_10_incidents_highest_casualties.png')
```
